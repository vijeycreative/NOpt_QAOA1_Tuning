```python
import gzip
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Function to save a dictionary to a gzip file
def save_dict_gzip(data_dict, filename):
    with gzip.open(filename, 'wb') as f:
        pickle.dump(data_dict, f, protocol=pickle.HIGHEST_PROTOCOL)

# Function to load a dictionary from a gzip file
def load_dict_gzip(filename):
    with gzip.open(filename, 'rb') as f:
        data_dict = pickle.load(f)
    return data_dict
```
